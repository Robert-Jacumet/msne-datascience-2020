```python
import matplotlib.pyplot as plt
import numpy as np

import torch
import torchvision
import torchvision.datasets as datasets

# we use torchvision to download the dataset. root parameters says we store ot in a folder "data" and train
# is true since we get the trainset, we want to download it so True and no transform is applied. Only needs
#to be downloaded once. In addition we must transform the data that is made from PIL images
#to Tensors by choosing the following transform
transform = torchvision.transforms.ToTensor()

mnist_trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform) 

#We now download the Testset
mnist_testset= datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Jsut to get a feel:
print(len(mnist_trainset))
print(len(mnist_testset))

# until here we just initialized our dataset
##########################################
#Now DataLoading. We can sample batches and each Batch is a dict of size batch_size
#that stores a label and the image

Batch_size=4

#Data loaader provides a sampler to iterate over the dataset
trainloader = torch.utils.data.DataLoader(mnist_trainset, batch_size=Batch_size, shuffle=True)
testloader = torch.utils.data.DataLoader(mnist_testset, batch_size=Batch_size, shuffle=False)

###########################################
#Data visualization (Just to get a better feel for our Dataset)

def show_batch(batch):
    im = torchvision.utils.make_grid(batch)
    plt.imshow(np.transpose(im.numpy(),(1,2,0))) #we take the second and third dimension as x and y and 
    # the first last, beacuse it gives us only the number of the pic in batch
    
    
#iter    
dataiter= iter(trainloader)
images, labels = dataiter.next()

print('Labels: ', labels)
print('Batch shape: ', images.size())
show_batch(images)
```
